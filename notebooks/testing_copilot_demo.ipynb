## 2. Gerar Testes UnitÃ¡rios

Vamos criar testes para uma funÃ§Ã£o simples de calculadora.


```python
# CÃ³digo de exemplo - Calculadora
codigo_calculadora = '''
class Calculadora:
    """Calculadora simples com operaÃ§Ãµes bÃ¡sicas."""
    
    def somar(self, a: float, b: float) -> float:
        """Soma dois nÃºmeros."""
        return a + b
    
    def subtrair(self, a: float, b: float) -> float:
        """Subtrai b de a."""
        return a - b
    
    def multiplicar(self, a: float, b: float) -> float:
        """Multiplica dois nÃºmeros."""
        return a * b
    
    def dividir(self, a: float, b: float) -> float:
        """Divide a por b."""
        if b == 0:
            raise ValueError("DivisÃ£o por zero nÃ£o Ã© permitida")
        return a / b
'''

print("ðŸ“ CÃ³digo da Calculadora:")
print(codigo_calculadora)
```

```python
# Gerar testes unitÃ¡rios com pytest
print("ðŸ§ª Gerando testes unitÃ¡rios...\n")

resultado = testing.generate_unit_tests(
    code=codigo_calculadora,
    language="python",
    framework="pytest"
)

if resultado.success:
    print("âœ… Testes gerados com sucesso!\n")
    print(resultado.content)
else:
    print(f"âŒ Erro: {resultado.content}")
```

## 3. Testar via API (Docker)

Se o Docker estiver rodando, podemos testar via API HTTP.


```python
import httpx

API_URL = "http://localhost:8000"

# Verificar se API estÃ¡ disponÃ­vel
try:
    response = httpx.get(f"{API_URL}/health", timeout=5)
    print(f"âœ… API disponÃ­vel: {response.json()}")
    API_DISPONIVEL = True
except:
    print("âš ï¸ API nÃ£o estÃ¡ rodando. Execute: make docker-up")
    API_DISPONIVEL = False
```

```python
# Gerar testes via API
if API_DISPONIVEL:
    codigo_simples = '''
def fibonacci(n: int) -> int:
    """Calcula o n-Ã©simo nÃºmero de Fibonacci."""
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

def fatorial(n: int) -> int:
    """Calcula o fatorial de n."""
    if n < 0:
        raise ValueError("Fatorial nÃ£o definido para negativos")
    if n <= 1:
        return 1
    return n * fatorial(n-1)
'''
    
    print("ðŸ§ª Gerando testes via API...\n")
    
    response = httpx.post(
        f"{API_URL}/api/test",
        json={
            "code": codigo_simples,
            "language": "python",
            "framework": "pytest"
        },
        timeout=120
    )
    
    if response.status_code == 200:
        resultado = response.json()
        print("âœ… Testes gerados via API:\n")
        print(resultado.get("content", resultado))
    else:
        print(f"âŒ Erro: {response.text}")
else:
    print("âš ï¸ Pulando teste via API (nÃ£o disponÃ­vel)")
```
